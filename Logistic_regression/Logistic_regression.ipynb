```python
import csv
```

```python
with open("/content/drive/MyDrive/Hackathon/dataset1.csv") as file:
  reader = csv.reader(file, delimiter=',')
  X_tmp =[]
  y_tmp =[]

  for line in reader:
    X_tmp.append(line[0].lower().strip())
    count = count +1
    #new = line[1].split("|")
    #print(new)
    y_tmp.append(line[1].lower().strip())


print(len(X_tmp))
print(len(y_tmp))

```

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X_txt_train, X_txt_test, y_train_text, y_test_text = train_test_split(X_tmp, y_tmp, test_size=0.2, random_state=42)
```

```python
X_train = np.array(X_txt_train)
X_test = np.array(X_txt_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vec =TfidfVectorizer(analyzer='word' , stop_words='english',)

X_train = vec.fit_transform(X_txt_train) # This should be a matrix
X_test = vec.transform(X_txt_test)   # This should be a matrix
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train, y_train_text)
y_pred = reg.predict(X_test)
```

```python
print(X_txt_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, f1_score
```

```python

accuracy_score(y_test_text,y_pred)
```

```python

f1 = f1_score(y_test_text,y_pred, average="macro")


f1 = f1_score(y_test_text, y_pred, average="macro")

print("F1: {:.4f}".format(f1))
```

```python
from sklearn.metrics import precision_score,recall_score

precision = precision_score(y_pred, y_test_text,average="macro")
recall = recall_score(y_pred, y_test_text, average="macro")

print(precision)
print(recall)
```
